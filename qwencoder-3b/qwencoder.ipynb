```python
import json

def convert_dataset(file_path, output_path):
    # Đọc file gốc
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    new_data = []

    for item in data:
        probing_input = item.get("probing input", "")
        deprecated_apis = item.get("deprecated api", [])
        # replacement_api = item.get("replacement api", "")

        # Nếu deprecated_api là list → tách từng cái
        for dep_api in deprecated_apis:
            new_sample = {
                "input": probing_input,
                "deprecated_api": dep_api,
                # "replacement_api": replacement_api
            }
            new_data.append(new_sample)

    # Lưu file mới
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(new_data, f, indent=2, ensure_ascii=False)


# ====== dùng ======
file_path = "D:\\EDAPI\\EDAPI-Bench-main\\data\\EditDeprecatedAPI\\qwencoder-3b\\all.json"          # đường dẫn file input
output_path = "D:\\EDAPI\\EDAPI-Bench-main\\data\\EditDeprecatedAPI\\qwencoder-3b\\qwencoder.json" # file output

convert_dataset(file_path, output_path)
```
