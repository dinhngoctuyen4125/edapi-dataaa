```python
import json
import os

# Đường dẫn file gốc (đã gộp nhưng chưa lọc trùng)
input_path = r"D:\EDAPI\EDAPI-Bench-main\data\EditDeprecatedAPI\merged_solve.json"

# Đường dẫn file mới (sau khi đã lọc trùng)
output_path = r"D:\EDAPI\EDAPI-Bench-main\data\EditDeprecatedAPI\cleaned_merged_solve.json"

def remove_duplicates_to_new_file(in_path, out_path):
    if not os.path.exists(in_path):
        print(f"Lỗi: Không tìm thấy file gốc '{in_path}'")
        return

    # 1. Đọc dữ liệu từ file gốc
    with open(in_path, 'r', encoding='utf-8') as f:
        try:
            data = json.load(f)
        except json.JSONDecodeError:
            print("Lỗi: File gốc không đúng định dạng JSON.")
            return

    print(f"Số lượng mẫu ban đầu: {len(data)}")

    # 2. Xử lý lọc trùng
    unique_data = []
    seen = set()

    for item in data:
        # Chuyển dictionary thành chuỗi để kiểm tra trùng lặp
        item_str = json.dumps(item, sort_keys=True)
        
        if item_str not in seen:
            seen.add(item_str)
            unique_data.append(item)

    duplicates_count = len(data) - len(unique_data)

    # 3. Luôn ghi ra file mới (dù có mẫu trùng hay không)
    os.makedirs(os.path.dirname(out_path), exist_ok=True)
    with open(out_path, 'w', encoding='utf-8') as f:
        json.dump(unique_data, f, indent=4, ensure_ascii=False)
        
    print("-" * 40)
    if duplicates_count > 0:
        print(f"Đã phát hiện và loại bỏ: {duplicates_count} mẫu trùng lặp.")
    else:
        print("Không có mẫu nào trùng lặp.")
        
    print(f"Số lượng mẫu giữ lại: {len(unique_data)}")
    print(f"Đã lưu thành công vào file MỚI tại: {out_path}")

# Chạy hàm
remove_duplicates_to_new_file(input_path, output_path)
```
