```python
import json
import os

# 1. Khai báo danh sách đường dẫn đầu vào
paths = [
    r"D:\EDAPI\EDAPI-Bench-main\data\EditDeprecatedAPI\codegemma-2b\solve.json",
    r"D:\EDAPI\EDAPI-Bench-main\data\EditDeprecatedAPI\deepseek-1.3b\solve.json",
    r"D:\EDAPI\EDAPI-Bench-main\data\EditDeprecatedAPI\qwencoder-3b\solve.json"
]

# 2. Khai báo đường dẫn đầu ra
output_path = r"D:\EDAPI\EDAPI-Bench-main\data\EditDeprecatedAPI\merged_solve.json"

merged_list = []

# 3. Đọc và gộp dữ liệu
for path in paths:
    if os.path.exists(path):
        with open(path, 'r', encoding='utf-8') as f:
            try:
                data = json.load(f)
                # Đảm bảo dữ liệu đọc được là một list trước khi gộp
                if isinstance(data, list):
                    merged_list.extend(data)
                    print(f"Đã nạp {len(data)} mẫu từ: {os.path.basename(path)}")
                else:
                    print(f"Cảnh báo: Dữ liệu trong '{os.path.basename(path)}' không phải là một list.")
            except json.JSONDecodeError:
                print(f"Lỗi: File '{os.path.basename(path)}' không đúng định dạng JSON.")
    else:
        print(f"Lỗi: Không tìm thấy file tại '{path}'")

# 4. Lưu kết quả ra file mới
# Đảm bảo thư mục chứa file đầu ra đã tồn tại, nếu chưa thì tự động tạo
os.makedirs(os.path.dirname(output_path), exist_ok=True)

with open(output_path, 'w', encoding='utf-8') as f:
    # indent=4 giúp file json output dễ đọc (pretty-print)
    # ensure_ascii=False để hiển thị đúng các ký tự Unicode nếu có
    json.dump(merged_list, f, indent=4, ensure_ascii=False)

print("-" * 40)
print(f"Hoàn tất! Tổng cộng đã gộp: {len(merged_list)} mẫu.")
print(f"File gộp được lưu tại: {output_path}")
```
